```python
df1 = pd.read_csv('steamdbCleanedPT3.csv')
feature_columns = {'Action','Indie','Strategy','RPG','Casual','Simulation','Racing','Adventure','Nudity','Free to Play','Sports','Violent','Massively Multiplayer','Early Access','Sexual Content','Gore'}
feature_counts = {}
for column in feature_columns:
    counts = df1[column].value_counts()
    feature_counts[column] = counts
total_samples = len(df1) 
total_percent = 0

for feature, counts in feature_counts.items():
    print("Feature:", feature)
    
    count_ones = counts[1] if 1 in counts else 0
    percentage_ones = count_ones / total_samples * 100
    raw_percent = count_ones/total_samples
    squared_percent = raw_percent * raw_percent
    total_percent += squared_percent

    print(f"Count of 1s: {count_ones}")
    print(f"Percentage of 1s: {percentage_ones:.2f}%")
    print(f"Raw percent: {raw_percent}")
    print(f"{squared_percent}\n")

print(f"Baseline accuracy: {total_percent}")

    

```
